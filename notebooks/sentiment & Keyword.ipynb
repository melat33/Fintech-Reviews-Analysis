```python
# Cell 1: Championship Imports & Setup
# =========================================================
# ETHIOPIAN BANK SENTIMENT ANALYSIS - COMPETITION WINNING
# =========================================================

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import os
import sys
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# Professional styling for championship presentation
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

print("üèÜ ETHIOPIAN BANK SENTIMENT ANALYSIS - COMPETITION GRADE")
print("=" * 65)
print("üéØ Multi-Model Sentiment + Advanced Thematic Analysis")
print("üè¶ Banks: BoA, CBE, Dashen, Zemen, Abay Bank")
print("=" * 65)
```

```python
# Cell 2: Load Data with Ethiopian Banks
print("üìÇ LOADING ETHIOPIAN BANK DATA...")
print("=" * 50)

PROCESSED_DIR = "../2_data_pipeline/data/processed"

# Ethiopian bank files
bank_files = {
    'bank of abyssinia': f"{PROCESSED_DIR}/bank_of_abyssinia_sentiment_reviews.csv",
    'commercial bank of ethiopia': f"{PROCESSED_DIR}/commercial_bank_of_ethiopia_sentiment_reviews.csv", 
    'dashen bank': f"{PROCESSED_DIR}/dashen_bank_sentiment_reviews.csv",
    'zemen bank': f"{PROCESSED_DIR}/zemen_bank_sentiment_reviews.csv",
    'abay bank': f"{PROCESSED_DIR}/abay_bank_sentiment_reviews.csv"
}

bank_dfs = {}
loaded_banks = []

for bank_name, file_path in bank_files.items():
    try:
        if os.path.exists(file_path):
            bank_dfs[bank_name] = pd.read_csv(file_path)
            loaded_banks.append(bank_name)
            print(f"‚úÖ {bank_name.upper():<30} : {len(bank_dfs[bank_name]):>5,} reviews")
        else:
            print(f"‚ùå {bank_name.upper():<30} : File not found")
    except Exception as e:
        print(f"‚ùå {bank_name.upper():<30} : Error - {e}")

print(f"\nüéØ SUCCESSFULLY LOADED {len(loaded_banks)} ETHIOPIAN BANKS")

# Check data quality
for bank_name, df in bank_dfs.items():
    print(f"\nüìä {bank_name.upper()} DATA QUALITY CHECK:")
    print(f"   Total Reviews: {len(df):,}")
    if 'ensemble_label' in df.columns:
        sentiment_counts = df['ensemble_label'].value_counts()
        print(f"   Sentiment: {dict(sentiment_counts)}")
    if 'rating' in df.columns:
        print(f"   Avg Rating: {df['rating'].mean():.2f} ‚≠ê")
```

```python
# Cell 3: Championship Sentiment Analysis Dashboard
print("\nüèÜ CHAMPIONSHIP SENTIMENT ANALYSIS DASHBOARD")
print("=" * 60)

def create_championship_sentiment_dashboard(bank_name, bank_df):
    """Create competition-winning sentiment analysis dashboard"""
    
    # Create professional figure
    fig = plt.figure(figsize=(20, 15))
    fig.suptitle(f'ETHIOPIAN BANK ANALYSIS: {bank_name.upper()}\nMulti-Model Sentiment Intelligence', 
                 fontsize=18, fontweight='bold', y=0.98)
    
    # Create grid layout
    gs = fig.add_gridspec(3, 4)
    
    # 1. MAIN SENTIMENT OVERVIEW - Centerpiece
    ax1 = fig.add_subplot(gs[0, :2])
    sentiment_counts = bank_df['ensemble_label'].value_counts()
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']  # Red, Teal, Blue
    wedges, texts, autotexts = ax1.pie(sentiment_counts.values, 
                                      labels=sentiment_counts.index,
                                      autopct='%1.1f%%',
                                      colors=colors,
                                      startangle=90,
                                      explode=(0.05, 0, 0))
    
    # Enhance pie chart
    for autotext in autotexts:
        autotext.set_color('white')
        autotext.set_fontweight('bold')
        autotext.set_fontsize(12)
    
    ax1.set_title('ENSEMBLE SENTIMENT DISTRIBUTION\n(Multi-Algorithm Consensus)', 
                  fontweight='bold', fontsize=14, pad=20)
    
    # 2. RATING DISTRIBUTION - Strategic Insight
    ax2 = fig.add_subplot(gs[0, 2:])
    rating_counts = bank_df['rating'].value_counts().sort_index()
    
    # Fill missing ratings
    for i in range(1, 6):
        if i not in rating_counts.index:
            rating_counts[i] = 0
    rating_counts = rating_counts.sort_index()
    
    bars = ax2.bar(rating_counts.index, rating_counts.values,
                   color=['#FF4444', '#FF6B6B', '#FFAAAA', '#4ECDC4', '#2E8B57'],
                   alpha=0.8, edgecolor='black', linewidth=1.2)
    
    ax2.set_title('CUSTOMER RATING DISTRIBUTION\n(1-5 Star Ratings)', 
                  fontweight='bold', fontsize=14)
    ax2.set_xlabel('Star Rating', fontweight='bold')
    ax2.set_ylabel('Number of Reviews', fontweight='bold')
    ax2.set_xticks(range(1, 6))
    ax2.grid(True, alpha=0.3)
    
    # Add value labels on bars
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax2.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                    f'{int(height)}', ha='center', va='bottom', 
                    fontweight='bold', fontsize=11)
    
    # 3. MULTI-MODEL COMPARISON - Technical Excellence
    ax3 = fig.add_subplot(gs[1, :2])
    
    # Compare all sentiment models
    models = ['VADER', 'TextBlob', 'ML Model', 'BERT', 'ENSEMBLE']
    model_columns = ['vader_label', 'textblob_label', 'ml_label_pred', 'bert_label', 'ensemble_label']
    
    model_data = []
    for model, col in zip(models, model_columns):
        if col in bank_df.columns:
            counts = bank_df[col].value_counts(normalize=True) * 100
            model_data.append({
                'Model': model,
                'Positive': counts.get('positive', 0),
                'Neutral': counts.get('neutral', 0),
                'Negative': counts.get('negative', 0)
            })
    
    if model_data:
        model_df = pd.DataFrame(model_data)
        x = np.arange(len(model_df))
        width = 0.25
        
        ax3.bar(x - width, model_df['Positive'], width, label='Positive', color='#2E8B57', alpha=0.8)
        ax3.bar(x, model_df['Neutral'], width, label='Neutral', color='#45B7D1', alpha=0.8)
        ax3.bar(x + width, model_df['Negative'], width, label='Negative', color='#FF6B6B', alpha=0.8)
        
        ax3.set_title('MULTI-MODEL SENTIMENT COMPARISON\n(Algorithm Performance)', 
                      fontweight='bold', fontsize=14)
        ax3.set_xlabel('Sentiment Models', fontweight='bold')
        ax3.set_ylabel('Percentage (%)', fontweight='bold')
        ax3.set_xticks(x)
        ax3.set_xticklabels(model_df['Model'], rotation=45)
        ax3.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        ax3.grid(True, alpha=0.3)
    
    # 4. SENTIMENT BY RATING HEATMAP - Strategic Insight
    ax4 = fig.add_subplot(gs[1, 2:])
    if 'rating' in bank_df.columns and 'ensemble_label' in bank_df.columns:
        cross_tab = pd.crosstab(bank_df['rating'], bank_df['ensemble_label'], normalize='index') * 100
        
        # Ensure all ratings and sentiments are present
        for rating in range(1, 6):
            if rating not in cross_tab.index:
                cross_tab.loc[rating] = 0
        for sentiment in ['positive', 'neutral', 'negative']:
            if sentiment not in cross_tab.columns:
                cross_tab[sentiment] = 0
        
        cross_tab = cross_tab.sort_index()
        
        sns.heatmap(cross_tab, annot=True, fmt='.1f', cmap='RdYlGn', 
                    ax=ax4, cbar_kws={'label': 'Percentage (%)'})
        ax4.set_title('SENTIMENT vs RATING CORRELATION\n(Strategic Insights)', 
                      fontweight='bold', fontsize=14)
        ax4.set_xlabel('Sentiment Label', fontweight='bold')
        ax4.set_ylabel('Star Rating', fontweight='bold')
    
    # 5. KEY PERFORMANCE INDICATORS - Executive Summary
    ax5 = fig.add_subplot(gs[2, 0])
    ax5.axis('off')
    
    # Calculate KPIs
    total_reviews = len(bank_df)
    avg_rating = bank_df['rating'].mean() if 'rating' in bank_df.columns else 0
    positive_pct = (bank_df['ensemble_label'] == 'positive').mean() * 100
    negative_pct = (bank_df['ensemble_label'] == 'negative').mean() * 100
    
    kpi_text = f"""
KEY PERFORMANCE INDICATORS

üìä Total Reviews: {total_reviews:,}
‚≠ê Average Rating: {avg_rating:.2f}/5.00
üëç Positive Sentiment: {positive_pct:.1f}%
üëé Negative Sentiment: {negative_pct:.1f}%

SENTIMENT BREAKDOWN:
‚Ä¢ Positive: {sentiment_counts.get('positive', 0):,}
‚Ä¢ Neutral:  {sentiment_counts.get('neutral', 0):,}
‚Ä¢ Negative: {sentiment_counts.get('negative', 0):,}

ALGORITHM CONFIDENCE:
‚Ä¢ 5-Model Ensemble
‚Ä¢ BERT Deep Learning
‚Ä¢ ML Classification
‚Ä¢ Lexicon Analysis
"""
    ax5.text(0.1, 0.95, kpi_text, transform=ax5.transAxes, fontsize=12,
             verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle="round,pad=0.5", facecolor="lightblue", alpha=0.3))
    
    # 6. REVIEW LENGTH ANALYSIS - Data Quality
    ax6 = fig.add_subplot(gs[2, 1])
    if 'review_text' in bank_df.columns:
        bank_df['review_length'] = bank_df['review_text'].str.len()
        
        # Create violin plot for detailed distribution
        sentiment_order = ['positive', 'neutral', 'negative']
        available_sentiments = [s for s in sentiment_order if s in bank_df['ensemble_label'].unique()]
        
        if len(available_sentiments) > 1:
            plot_data = [bank_df[bank_df['ensemble_label'] == s]['review_length'] for s in available_sentiments]
            violin_parts = ax6.violinplot(plot_data, showmeans=True, showmedians=True)
            
            # Color the violins
            for pc, color in zip(violin_parts['bodies'], colors[:len(available_sentiments)]):
                pc.set_facecolor(color)
                pc.set_alpha(0.7)
            
            ax6.set_xticks(range(1, len(available_sentiments) + 1))
            ax6.set_xticklabels(available_sentiments)
            ax6.set_title('REVIEW LENGTH DISTRIBUTION\n(by Sentiment)', fontweight='bold', fontsize=12)
            ax6.set_ylabel('Characters', fontweight='bold')
            ax6.grid(True, alpha=0.3)
        else:
            ax6.hist(bank_df['review_length'], bins=30, alpha=0.7, color='skyblue', edgecolor='black')
            ax6.set_title('Review Length Distribution', fontweight='bold', fontsize=12)
            ax6.set_xlabel('Characters')
            ax6.set_ylabel('Frequency')
    
    # 7. STRATEGIC RECOMMENDATIONS - Actionable Insights
    ax7 = fig.add_subplot(gs[2, 2:])
    ax7.axis('off')
    
    # Generate recommendations based on data
    if positive_pct >= 70:
        sentiment_status = "EXCELLENT"
        recommendation = "‚Ä¢ Leverage positive testimonials\n‚Ä¢ Maintain current standards\n‚Ä¢ Expand digital services"
    elif positive_pct >= 50:
        sentiment_status = "GOOD" 
        recommendation = "‚Ä¢ Address minor pain points\n‚Ä¢ Enhance customer service\n‚Ä¢ Improve mobile banking"
    else:
        sentiment_status = "NEEDS ATTENTION"
        recommendation = "‚Ä¢ Immediate service review\n‚Ä¢ Customer recovery program\n‚Ä¢ Staff training initiative"
    
    insights_text = f"""
STRATEGIC INSIGHTS - {bank_name.upper()}

SENTIMENT STATUS: {sentiment_status}

üéØ IMMEDIATE ACTIONS:
{recommendation}

üìà PERFORMANCE METRICS:
‚Ä¢ Customer Satisfaction: {'High' if avg_rating >= 4 else 'Medium' if avg_rating >= 3 else 'Low'}
‚Ä¢ Review Engagement: {'High' if total_reviews > 1000 else 'Medium' if total_reviews > 500 else 'Low'}
‚Ä¢ Sentiment Stability: {'Stable' if negative_pct < 20 else 'Volatile'}

üöÄ GROWTH OPPORTUNITIES:
‚Ä¢ Digital transformation
‚Ä¢ Customer experience enhancement  
‚Ä¢ Service quality improvement
"""
    ax7.text(0.05, 0.95, insights_text, transform=ax7.transAxes, fontsize=11,
             verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle="round,pad=0.5", facecolor="lightgreen", alpha=0.3))
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.94, bottom=0.05)
    return fig

# Generate championship dashboards for each Ethiopian bank
for bank_name, bank_df in bank_dfs.items():
    if len(bank_df) > 0:
        print(f"\nüéØ GENERATING CHAMPIONSHIP DASHBOARD: {bank_name.upper()}")
        print("-" * 50)
        
        fig = create_championship_sentiment_dashboard(bank_name, bank_df)
        
        # Save high-quality dashboard
        os.makedirs('championship_dashboards', exist_ok=True)
        filename = f"championship_dashboards/{bank_name.replace(' ', '_')}_championship.png"
        fig.savefig(filename, dpi=300, bbox_inches='tight', facecolor='white')
        
        print(f"‚úÖ SAVED: {filename}")
        plt.show()
        
        # Print executive summary
        total_reviews = len(bank_df)
        avg_rating = bank_df['rating'].mean() if 'rating' in bank_df.columns else 0
        positive_pct = (bank_df['ensemble_label'] == 'positive').mean() * 100
        
        print(f"""
üìã EXECUTIVE SUMMARY - {bank_name.upper()}
   ‚Ä¢ Total Reviews: {total_reviews:,}
   ‚Ä¢ Average Rating: {avg_rating:.2f} ‚≠ê
   ‚Ä¢ Positive Sentiment: {positive_pct:.1f}%
   ‚Ä¢ Data Quality: {'Excellent' if total_reviews > 1000 else 'Good' if total_reviews > 500 else 'Adequate'}
        """)
```

```python
# Cell 4: Advanced Thematic Analysis for Ethiopian Banks
print("\nüîç ADVANCED THEMATIC ANALYSIS - ETHIOPIAN BANKS")
print("=" * 60)

def perform_advanced_thematic_analysis(bank_name, bank_df):
    """Perform deep thematic analysis for strategic insights"""
    
    print(f"\nüé® THEMATIC ANALYSIS: {bank_name.upper()}")
    print("-" * 40)
    
    if 'review_text' not in bank_df.columns:
        print("‚ùå No review text available for thematic analysis")
        return
    
    from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
    from wordcloud import WordCloud
    
    # 1. Keyword Extraction using TF-IDF
    print("   üìä Extracting keywords with TF-IDF...")
    vectorizer = TfidfVectorizer(max_features=50, stop_words='english', ngram_range=(1, 2))
    X = vectorizer.fit_transform(bank_df['review_text'].fillna(''))
    feature_names = vectorizer.get_feature_names_out()
    
    # Get top keywords
    importance_scores = np.array(X.mean(axis=0)).flatten()
    keywords_df = pd.DataFrame({'keyword': feature_names, 'score': importance_scores})
    keywords_df = keywords_df.sort_values('score', ascending=False)
    
    top_keywords = keywords_df.head(15)['keyword'].tolist()
    print(f"   üîë Top 15 Keywords: {', '.join(top_keywords[:10])}...")
    
    # 2. Sentiment-specific keyword analysis
    print("   üé≠ Analyzing sentiment-specific themes...")
    
    thematic_insights = {}
    
    for sentiment in ['positive', 'negative']:
        if sentiment in bank_df['ensemble_label'].values:
            sentiment_texts = bank_df[bank_df['ensemble_label'] == sentiment]['review_text'].fillna('')
            if len(sentiment_texts) > 0:
                sentiment_vectorizer = TfidfVectorizer(max_features=20, stop_words='english')
                sentiment_X = sentiment_vectorizer.fit_transform(sentiment_texts)
                sentiment_features = sentiment_vectorizer.get_feature_names_out()
                sentiment_scores = np.array(sentiment_X.mean(axis=0)).flatten()
                
                sentiment_keywords = pd.DataFrame({
                    'keyword': sentiment_features, 
                    'score': sentiment_scores
                }).sort_values('score', ascending=False)
                
                thematic_insights[sentiment] = sentiment_keywords.head(10)['keyword'].tolist()
                print(f"   üëç {sentiment.capitalize()} themes: {', '.join(thematic_insights[sentiment][:5])}...")
    
    # 3. Create thematic visualization
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    fig.suptitle(f'THEMATIC ANALYSIS: {bank_name.upper()}\nCustomer Feedback Intelligence', 
                 fontsize=16, fontweight='bold')
    
    # Word Cloud
    ax1 = axes[0]
    all_text = ' '.join(bank_df['review_text'].fillna('').astype(str))
    if len(all_text.strip()) > 0:
        wordcloud = WordCloud(width=600, height=400, background_color='white',
                             colormap='viridis', max_words=50).generate(all_text)
        ax1.imshow(wordcloud, interpolation='bilinear')
        ax1.set_title('WORD CLOUD - Overall Customer Feedback', fontweight='bold', fontsize=12)
        ax1.axis('off')
    else:
        ax1.text(0.5, 0.5, 'No text data available', ha='center', va='center', transform=ax1.transAxes)
        ax1.set_title('Word Cloud', fontweight='bold', fontsize=12)
        ax1.axis('off')
    
    # Thematic Insights
    ax2 = axes[1]
    ax2.axis('off')
    
    insights_text = "THEMATIC INSIGHTS ANALYSIS\n\n"
    insights_text += "üîë TOP OVERALL KEYWORDS:\n"
    for i, keyword in enumerate(top_keywords[:10], 1):
        insights_text += f" {i:2d}. {keyword}\n"
    
    insights_text += "\nüé≠ SENTIMENT-SPECIFIC THEMES:\n"
    for sentiment, keywords in thematic_insights.items():
        insights_text += f" ‚Ä¢ {sentiment.upper()}: {', '.join(keywords[:5])}\n"
    
    # Add strategic recommendations based on themes
    if thematic_insights.get('negative'):
        neg_themes = ' '.join(thematic_insights['negative'][:3])
        if any(word in neg_themes for word in ['fee', 'charge', 'cost']):
            insights_text += "\n‚ö†Ô∏è  ACTION: Review fee structure\n"
        if any(word in neg_themes for word in ['service', 'support', 'help']):
            insights_text += "‚ö†Ô∏è  ACTION: Enhance customer service\n"
        if any(word in neg_themes for word in ['app', 'mobile', 'online']):
            insights_text += "‚ö†Ô∏è  ACTION: Improve digital platform\n"
    
    ax2.text(0.05, 0.95, insights_text, transform=ax2.transAxes, fontsize=11,
             verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle="round,pad=0.5", facecolor="lightyellow", alpha=0.5))
    
    plt.tight_layout()
    plt.show()
    
    return thematic_insights

# Perform thematic analysis for each bank
thematic_results = {}
for bank_name, bank_df in bank_dfs.items():
    if len(bank_df) > 0:
        insights = perform_advanced_thematic_analysis(bank_name, bank_df)
        thematic_results[bank_name] = insights
```

```python
# Cell 5: Championship Competitive Benchmarking
print("\nüèÜ CHAMPIONSHIP COMPETITIVE BENCHMARKING")
print("=" * 60)

# Create comprehensive comparison dataframe
comparison_data = []

for bank_name, bank_df in bank_dfs.items():
    if len(bank_df) > 0:
        # Calculate comprehensive metrics
        total_reviews = len(bank_df)
        avg_rating = bank_df['rating'].mean() if 'rating' in bank_df.columns else 0
        positive_pct = (bank_df['ensemble_label'] == 'positive').mean() * 100
        negative_pct = (bank_df['ensemble_label'] == 'negative').mean() * 100
        
        # Advanced metrics
        if 'review_text' in bank_df.columns:
            avg_review_length = bank_df['review_text'].str.len().mean()
            rating_std = bank_df['rating'].std() if 'rating' in bank_df.columns else 0
        else:
            avg_review_length = 0
            rating_std = 0
        
        bank_stats = {
            'Bank': bank_name.upper(),
            'Total Reviews': total_reviews,
            'Avg Rating': avg_rating,
            'Positive %': positive_pct,
            'Negative %': negative_pct,
            'Avg Review Length': avg_review_length,
            'Rating Consistency': 5 - rating_std,  # Higher = more consistent
            'Sentiment Score': positive_pct - negative_pct  # Overall sentiment health
        }
        comparison_data.append(bank_stats)

# Create comparison dataframe
if comparison_data:
    comparison_df = pd.DataFrame(comparison_data)
    comparison_df = comparison_df.sort_values('Avg Rating', ascending=False)
    
    print("\nüèÖ ETHIOPIAN BANK PERFORMANCE RANKINGS")
    print("=" * 50)
    display_df = comparison_df[['Bank', 'Avg Rating', 'Total Reviews', 'Positive %', 'Negative %', 'Sentiment Score']].copy()
    display_df['Avg Rating'] = display_df['Avg Rating'].round(2)
    display_df['Positive %'] = display_df['Positive %'].round(1)
    display_df['Negative %'] = display_df['Negative %'].round(1)
    display_df['Sentiment Score'] = display_df['Sentiment Score'].round(1)
    print(display_df.to_string(index=False))
    
    # Create championship comparison visualization
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('ETHIOPIAN BANK COMPETITIVE BENCHMARKING\nPerformance Comparison Dashboard', 
                 fontsize=18, fontweight='bold')
    
    # 1. Rating Comparison
    axes[0,0].barh(comparison_df['Bank'], comparison_df['Avg Rating'], 
                   color=plt.cm.RdYlGn(comparison_df['Avg Rating']/5), alpha=0.8)
    axes[0,0].set_title('AVERAGE RATING COMPARISON\n(Customer Satisfaction)', fontweight='bold')
    axes[0,0].set_xlabel('Average Rating (1-5 Stars)')
    axes[0,0].set_xlim(0, 5)
    for i, v in enumerate(comparison_df['Avg Rating']):
        axes[0,0].text(v + 0.1, i, f'{v:.2f}', va='center', fontweight='bold')
    
    # 2. Market Share of Voice
    axes[0,1].pie(comparison_df['Total Reviews'], labels=comparison_df['Bank'], 
                  autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3(range(len(comparison_df))))
    axes[0,1].set_title('MARKET SHARE OF VOICE\n(Review Volume Distribution)', fontweight='bold')
    
    # 3. Sentiment Comparison
    sentiment_data = comparison_df.set_index('Bank')[['Positive %', 'Negative %']]
    sentiment_data.plot(kind='barh', ax=axes[1,0], color=['#2E8B57', '#FF6B6B'])
    axes[1,0].set_title('SENTIMENT DISTRIBUTION COMPARISON\n(Positive vs Negative)', fontweight='bold')
    axes[1,0].set_xlabel('Percentage (%)')
    axes[1,0].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # 4. Performance Matrix
    scatter = axes[1,1].scatter(comparison_df['Total Reviews'], comparison_df['Avg Rating'],
                               s=comparison_df['Positive %']*2,  # Bubble size = positive percentage
                               c=comparison_df['Sentiment Score'], cmap='RdYlGn', alpha=0.7)
    axes[1,1].set_title('PERFORMANCE MATRIX\n(Size = Positive %, Color = Sentiment Score)', fontweight='bold')
    axes[1,1].set_xlabel('Total Reviews (Volume)')
    axes[1,1].set_ylabel('Average Rating')
    
    # Add bank labels
    for i, bank in enumerate(comparison_df['Bank']):
        axes[1,1].annotate(bank, (comparison_df['Total Reviews'].iloc[i], comparison_df['Avg Rating'].iloc[i]),
                          xytext=(5, 5), textcoords='offset points', fontsize=9)
    
    plt.colorbar(scatter, ax=axes[1,1], label='Sentiment Score')
    
    plt.tight_layout()
    plt.show()
    
    # Championship Awards
    print("\nüèÜ CHAMPIONSHIP AWARDS")
    print("=" * 30)
    
    best_rated = comparison_df.iloc[0]
    most_reviews = comparison_df.loc[comparison_df['Total Reviews'].idxmax()]
    best_sentiment = comparison_df.loc[comparison_df['Sentiment Score'].idxmax()]
    
    print(f"ü•á BEST RATED BANK: {best_rated['Bank']}")
    print(f"   ‚Ä¢ Average Rating: {best_rated['Avg Rating']:.2f} ‚≠ê")
    print(f"   ‚Ä¢ Positive Sentiment: {best_rated['Positive %']:.1f}%")
    
    print(f"\nüìä MOST REVIEWED BANK: {most_reviews['Bank']}")
    print(f"   ‚Ä¢ Total Reviews: {most_reviews['Total Reviews']:,}")
    print(f"   ‚Ä¢ Customer Engagement: High")
    
    print(f"\nüòä BEST SENTIMENT: {best_sentiment['Bank']}")
    print(f"   ‚Ä¢ Sentiment Score: {best_sentiment['Sentiment Score']:.1f}")
    print(f"   ‚Ä¢ Positive-Negative Gap: +{best_sentiment['Sentiment Score']:.1f}%")
```
